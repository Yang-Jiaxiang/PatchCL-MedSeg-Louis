```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readCSV(path):
    df = pd.read_csv(path)
    return df
```

```python
def plotCurve(df, weight = 0.1):
    plt.figure(figsize=(15, 15))

    # 創建 2x2 的圖表佈局
    fig, axs = plt.subplots(3, 2, figsize=(15, 15))

    # 第一張圖 - 監督損失
    axs[0, 0].plot(df['t_loss'], label='Training')
    axs[0, 0].plot(df['v_loss'], label='Validation')
    axs[0, 0].set_xlabel('Epoch')
    axs[0, 0].set_ylabel('Supervised Loss')
    axs[0, 0].set_title('Training and Validation total(Segmentation + Contrastive) Loss Curve')
    axs[0, 0].legend()
    axs[0, 0].grid(True)
    axs[0, 0].set_ylim(0, weight*10)  # 設定 y 軸顯示範圍為 0 到 1
        
    # 第二張圖 - 對比損失
    axs[0, 1].plot(df['t_supervised_loss'], label='Training')
    axs[0, 1].plot(df['v_supervised_loss'], label='Validation')
    axs[0, 1].set_xlabel('Epoch')
    axs[0, 1].set_ylabel('Loss')
    axs[0, 1].set_title('Training and Validation Segmentation Loss Curve')
    axs[0, 1].legend()
    axs[0, 1].grid(True)
    axs[0, 1].set_ylim(0, 1)  # 設定 y 軸顯示範圍為 0 到 1

    # 第二張圖 - 對比損失
    axs[1, 0].plot(df['t_contrastive_loss'], label='Training')
    axs[1, 0].plot(df['v_contrastive_loss'], label='Validation')
    axs[1, 0].set_xlabel('Epoch')
    axs[1, 0].set_ylabel('Loss')
    axs[1, 0].set_title('Training and Validation Contrastive Loss Curve')
    axs[1, 0].legend()
    axs[1, 0].grid(True)
    axs[1, 0].set_ylim(0, 10)  # 設定 y 軸顯示範圍為 0 到 1

    # 第三張圖 - 一致性損失
    axs[1, 1].plot(df['t_accuracy'], label='Training')
    axs[1, 1].plot(df['v_accuracy'], label='Validation')
    axs[1, 1].set_xlabel('Epoch')
    axs[1, 1].set_ylabel('Accuracy')
    axs[1, 1].set_title('Training and Validation Accuracy Curve')
    axs[1, 1].legend()
    axs[1, 1].grid(True)
    axs[1, 1].set_ylim(0, 1)  # 設定 y 軸顯示範圍為 0 到 1
    
    # 第三張圖 - 一致性損失
    axs[2, 0].plot(df['t_dice'], label='Training')
    axs[2, 0].plot(df['v_dice'], label='Validation')
    axs[2, 0].set_xlabel('Epoch')
    axs[2, 0].set_ylabel('Miou')
    axs[2, 0].set_title('Training and Validation Dice Curve')
    axs[2, 0].legend()
    axs[2, 0].grid(True)
    axs[2, 0].set_ylim(0, 1)  # 設定 y 軸顯示範圍為 0 到 1


    # 第三張圖 - 一致性損失
    axs[2, 1].plot(df['t_miou'], label='Training')
    axs[2, 1].plot(df['v_miou'], label='Validation')
    axs[2, 1].set_xlabel('Epoch')
    axs[2, 1].set_ylabel('Miou')
    axs[2, 1].set_title('Training and Validation Miou Curve')
    axs[2, 1].legend()
    axs[2, 1].grid(True)
    axs[2, 1].set_ylim(0, 1)  # 設定 y 軸顯示範圍為 0 到 1
        
    # 調整佈局，防止標題和軸標籤重疊
    plt.tight_layout()

    # 顯示圖表
    plt.show()
```
